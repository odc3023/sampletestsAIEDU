```python
from datasets import load_dataset

semeval_dataset = load_dataset("Atomi/semeval_2013_task_7_beetle_5way")
```

```python
from sentence_transformers import SentenceTransformer, util
import torch

# Load a SentenceTransformer model for similarity scoring
model = SentenceTransformer('all-MiniLM-L6-v2')

def calculate_similarity(reference, student_answer):
    """
    Calculate semantic similarity between the reference answer and the student's answer
    using cosine similarity of embeddings.
    """
    # Generate embeddings
    ref_embedding = model.encode(reference, convert_to_tensor=True)
    student_embedding = model.encode(student_answer, convert_to_tensor=True)

    # Compute cosine similarity
    similarity_score = util.cos_sim(ref_embedding, student_embedding)
    return similarity_score.item() * 100  # Convert to percentage
```

```python
# Iterate through a subset of the dataset and calculate scores
results = []
for entry in semeval_dataset["train"][:10]:  # Adjust [:10] for more entries
    question = entry["question"]
    ref_answer = entry["reference_answer"]
    student_answer = entry["student_answer"]

    # Calculate the similarity score
    score = calculate_similarity(ref_answer, student_answer)

    # Append results
    results.append({
        "question": question,
        "reference_answer": ref_answer,
        "student_answer": student_answer,
        "score": round(score, 2)  # Round to 2 decimal places
    })

# Print results
for result in results:
    print(f"Question: {result['question']}")
    print(f"Reference Answer: {result['reference_answer']}")
    print(f"Student Answer: {result['student_answer']}")
    print(f"Score: {result['score']}%\n")
```

```python
# Align fields across the lists using the zip() function
results = []
for question, ref_answer, student_answer in zip(
    semeval_dataset["train"]["question"],
    semeval_dataset["train"]["reference_answer"],
    semeval_dataset["train"]["student_answer"]
):
    # Calculate similarity between reference and student answers
    score = calculate_similarity(ref_answer, student_answer)

    # Append results
    results.append({
        "question": question,
        "reference_answer": ref_answer,
        "student_answer": student_answer,
        "score": round(score, 2)
    })

# Display the results
for result in results[:5]:  # Display only the first 5 results
    print(result)
```

```python
import pandas as pd

results_df = pd.DataFrame(results)
results_df.to_csv("automated_scoring_results.csv", index=False)
```
